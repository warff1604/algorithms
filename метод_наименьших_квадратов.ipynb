```python
# импорт библиотек
import numpy as np
import matplotlib.pyplot as plt
```

```python
# инициализация начальных значений
N = 100     # число экспериментов
sigma = 3   # стандартное отклонение наблюдаемых значений
k = 0.5     # теоретическое  значение параметра k
b = 2       # теоретическое значение параметра b
```

```python
# диапазон значений x
x = np.array(range(N))
```

```python
# значения теоретической функции
f = np.array([k * _ + b for _ in range(N)])
```

```python
# моделирование результатов наблюдений
y = f + np.random.normal(loc=0, scale=sigma, size=N)
```

```python
# отображение экспериментальных данных
plt.scatter(x, y, s=2, c='red')
plt.grid(True)
plt.show()
```

```python
# вычисляем коэффициенты k и b по экспериментальным данным
mx = x.sum() / N
my = y.sum() / N
a2 = np.dot(x.T, x) / N
a11 = np.dot(x.T, y) / N

kk = (a11 - mx * my) / (a2 - mx ** 2)
bb = my - kk * mx
```

```python
# значения полученной аппроксимации
ff = np.array([kk * _  + bb for _ in range(N)])
```

```python
# визуализация обоих графиков
plt.scatter(x, y, s=2, c='red')
plt.grid(True)
plt.plot(f)
plt.plot(ff, c='red');
```
